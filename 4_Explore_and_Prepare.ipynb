```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("clean_apartments_zurich_feb.csv", index_col=0)
```

```python
df.head()
```

```python
sns.catplot(x="price", y="rooms", data=df, aspect=5)
plt.xlabel('Price', fontsize=16)
plt.xticks(rotation=90)
plt.ylabel('Rooms', fontsize=16)
plt.title('Price per # Rooms', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=8)
plt.show()
```

```python
sns.catplot(y="area", x="price", data=df, aspect=5)
plt.xlabel('price', fontsize=16);
plt.xticks(rotation=90)
plt.ylabel('area [m2]', fontsize=16);
plt.title('Price per # Area', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=8)
plt.show()
```

```python
df1=df.groupby('location')['price'].mean()
df1.plot(y="price", x="location", kind="bar")
plt.xlabel("Location", size=14)
plt.ylabel("Average Price ", size=14)
plt.title("Average price in each location", size=18)
plt.show()
```

```python
df1=df.groupby(['location'])['price'].mean().reset_index()
sns.catplot(x="location", y="price", data=df1, kind="bar", height=5, aspect=1.5)
plt.xlabel("Location", size=14)
plt.xticks(rotation=90)
plt.ylabel("Average Price ", size=14)
plt.title("Average price in each location", size=18)
plt.tight_layout()
plt.show()
```

```python
df2=df.groupby(['location'])['price'].max().reset_index()
sns.catplot(x="location", y="price", data=df2, kind="bar", height=5, aspect=1.5)
plt.xlabel("Location", size=14)
plt.xticks(rotation=90)
plt.ylabel("Average Price ", size=14)
plt.title("Average price in each location", size=18)
plt.tight_layout()
plt.show()
```

```python
grouped_data = df.groupby('location')['price'].agg(['max', 'mean']).reset_index()

# plot the data using seaborn
sns.barplot(data=grouped_data, x='location', y='max', color='blue', label='Max price')
sns.lineplot(data=grouped_data, x='location', y='mean', color='red', label='Mean price')
plt.xticks(rotation=90)
plt.show()
```
