```python
import folium
import pandas as pd
```

```python
# Import folium MarkerCluster plugin
from folium.plugins import MarkerCluster
# Import folium MousePostion plugin
from folium.plugins import MousePosition
# Import folium DivIcon plugin
from folium.features import DivIcon
```

```python
houses_df = pd.read_csv("clean_apartments_zurich_feb.csv")
postal_df = pd.read_excel("Postal Codes.xlsx")
```

```python
houses_df = houses_df[['house_id', 'location']]
```

```python
m = folium.Map(location = [47.373878, 8.545094], zoom_start = 13)
```

```python
postal_df.head()
```

```python
def add_markers(item, loc):
    folium.Marker([postal_df.loc[item, "latitude"], postal_df.loc[item, "longitude"]], popup = loc, tooltip = loc).add_to(m)
```

```python
for i in range(len(postal_df)):
    location = postal_df.loc[i, "location"]
    add_markers(i, location)
```
